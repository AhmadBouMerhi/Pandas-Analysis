```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located
csv_crime = "Resources/crime_incident_data2017.csv"
# Import the data into a Pandas DataFrame
csv_crime = pd.read_csv(csv_crime)
csv_crime.head()
```

```python
# look for missing values
csv_crime.count()
```

```python
# drop null rows
drop_null = csv_crime.dropna(how='any')
```

```python
# verify counts
drop_null.count()
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
drop_null["Offense Type"].value_counts()
```

```python
# Combine similar offenses
drop_null["Offense Type"] = drop_null["Offense Type"].replace(
    {"Commercial Sex Acts": "Prostitution", "Assisting or Promoting Prostitution": "Prostitution"})
drop_null
```

```python
# Create a new DataFrame that looks into a specific neighborhood
king_crime_df = drop_null.loc[drop_null["Neighborhood"] == "King"]
king_crime_df
```
